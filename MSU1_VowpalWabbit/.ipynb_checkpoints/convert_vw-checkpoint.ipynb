```python
import re
import string
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
from nltk import word_tokenize

stemmer = SnowballStemmer('english')
 
 
def stem_text(text):
    stop = stopwords.words('english') + list(string.punctuation)
    res = ' '.join([stemmer.stem(w) for w in re.sub('\W+', ' ',text).lower().split() if not w in stop])
    return res

def stem_text(text):
    stop = stopwords.words('english') + list(string.punctuation)
    res = ' '.join([stemmer.stem(w) for w in re.sub('\W+', ' ',text).lower().split() if not w in stop])
    return res 
 
print(stem_text('experience, 207 of selling a service or a technical product ideally'))
```

```python
out_name = 'data/train.vw'
arr = train.values[0:1000]
out_file = open(out_name, 'w')
ind = 0
for doc in arr:
    out_file.write( str(doc[-2]) + " '" + str(ind)
    #+' |a_Title' + stem_text(doc[0])
    + (' |b_Descr ' + stem_text(doc[1]) if doc[1] is not np.nan else '')
    + (' |c_LocRaw ' + stem_text(doc[2]) if doc[2] is not np.nan else '')
    + (' |d_LocNorm ' + stem_text(doc[3]) if doc[3] is not np.nan else '')
    + (' |e_ContrType ' + stem_text(doc[4]) if doc[4] is not np.nan else '') 
    + (' |f_ContrTime ' + stem_text(doc[5]) if doc[5] is not np.nan else '')
    + (' |g_Comp ' + stem_text(doc[6]) if doc[6] is not np.nan else '')
    + (' |h_Categ ' + stem_text(doc[7]) if doc[7] is not np.nan else '')
    + (' |i_SalRaw ' + stem_text(doc[8]) if out_name.split('.')[0] != 'test' and doc[8] is not np.nan else '')
    # not needed! +' |j_SalNorm' + stem_text(doc[9])
    + (' |k_SrcName ' + stem_text(doc[10]) if doc[10] is not np.nan else '')
    + '\n' )
    ind +=1
out_file.close()
```

```python
def conv_pred(file_name):
    file = open(file_name, 'r')
    l = file.readlines()
    file.close()
    file = open(file_name, 'w')
    for line in l:
        file.write(', '.join([line.split()[1], line.split()[0]]) + '\n')
    file.close()
```
